```python
sns.countplot(x='is_fraud', data=df)
plt.title("Fraud vs Non-Fraud")
plt.show()
```

```python
sns.histplot(df['amt'], bins=100)
plt.title("Transaction Amount Distribution")
plt.show()
```

```python
sns.countplot(x='gender', hue='is_fraud', data=df)
plt.title("Fraud by Gender")
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.countplot(x='category', hue='is_fraud', data=df)
plt.xticks(rotation=45)
plt.title("Fraud by Category")
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.countplot(x='state', hue='is_fraud', data=df)
plt.xticks(rotation=90)
plt.title("Fraud by State")
plt.show()
```

```python
df['trans_date_trans_time'] = pd.to_datetime(df['trans_date_trans_time'])
```

```python
df['hour'] = df['trans_date_trans_time'].dt.hour
df['day'] = df['trans_date_trans_time'].dt.day
df['month'] = df['trans_date_trans_time'].dt.month
```

```python
sns.countplot(x='hour', hue='is_fraud', data=df)
plt.title("Fraud by Hour")
plt.show()
```

```python
df['dob'] = pd.to_datetime(df['dob'])
df['age'] = (pd.Timestamp.now() - df['dob']).dt.days // 365
```

```python
sns.histplot(data=df, x='age', hue='is_fraud', bins=50)
plt.title("Age vs Fraud")
plt.show()
```

```python
from geopy.distance import geodesic

def calculate_distance(row):
    return geodesic((row['lat'], row['long']),
                    (row['merch_lat'], row['merch_long'])).km

df['distance'] = df.apply(calculate_distance, axis=1)
```

```python
sns.histplot(df['distance'], bins=100)
plt.title("Distance Distribution")
plt.show()
```

```python
sns.boxplot(x='is_fraud', y='distance', data=df)
plt.title("Distance vs Fraud")
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['int64', 'float64'])
```

```python
corr = numeric_df.corr()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(corr, cmap='coolwarm')
plt.title("Correlation Matrix (Numeric Features Only)")
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(corr[['is_fraud']].sort_values(by='is_fraud', ascending=False),
            annot=True, cmap='coolwarm')
plt.title("Correlation with Fraud")
plt.show()
```
